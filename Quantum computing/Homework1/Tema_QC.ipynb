```python
#ex2
#iii
from qiskit import *
from qiskit import Aer
import numpy as np
import sys
#Changing the simulator 
backend = Aer.get_backend('unitary_simulator')
#The circuit without measurement
circuit1=QuantumCircuit(2)
circuit1.h(1)
circuit1.cx(0,1)
circuit1.h(1)
job = execute(circuit1, backend)
result = job.result()
#get the unitary matrix from the result object
A=result.get_unitary(circuit1, decimals=3);
print("Matricea unitara a circuitului 1 este \n",A)
circuit1.measure_all()
print(circuit1.draw())
simulator=Aer.get_backend('qasm_simulator')
result=execute(circuit1,simulator,shots=1024).result()
counts=result.get_counts()
print("Rezultatul simularii este ",counts)
#Changing the simulator 
backend = Aer.get_backend('unitary_simulator')
circuit2=QuantumCircuit(2)
circuit2.cz(1,0)
job = execute(circuit2, backend)
result = job.result()
#get the unitary matrix from the result object
B=result.get_unitary(circuit2, decimals=3);
print("Matricea unitara a circuitului 2 este \n",B)
circuit2.measure_all()
print(circuit2.draw())
simulator=Aer.get_backend('qasm_simulator')
result=execute(circuit2,simulator,shots=1024).result()
counts=result.get_counts()
print("Rezultatul simularii este ",counts)
def egalitate(A,B):
    eps=sys.float_info.epsilon
    for i in range(4):
        for j in range(4):
            if((A[i][j]-B[i][j]) >= eps):
                return 0
    return 1
if(egalitate(A,B)==0):
    print("Circuitele nu sunt egale");
else:
    print("Circuitele sunt egale");
```

```python
#ex1
from qiskit import *
from qiskit.visualization import plot_histogram
circuit=QuantumCircuit(3)
initial_state=[0,1]
circuit.initialize(initial_state,0)
initial_state=[1,0]
circuit.initialize(initial_state,1)
circuit.initialize(initial_state,2)
circuit.h(1)
circuit.cx(1,2)
circuit.barrier()
circuit.x(0)
circuit.x(2)
circuit.measure_all()
print(circuit.draw())
simulator=Aer.get_backend('qasm_simulator')
result2=execute(circuit,simulator,shots=8000).result()
counts2=result2.get_counts()
print("Rezultatul simularii este ",counts2)
plot_histogram(counts2).savefig('fig1.png')
```
