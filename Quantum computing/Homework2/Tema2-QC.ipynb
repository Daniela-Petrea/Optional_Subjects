```python
#Ex1
#XGATE
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute, Aer
from qiskit.circuit.library import XGate
import numpy as np
import sys
backend = Aer.get_backend('unitary_simulator')
qreg = QuantumRegister(4) 
qc = QuantumCircuit(qreg)
circuit = XGate().control(3, ctrl_state='010')
qc.append(circuit, [0, 1, 2, 3])
job = execute(qc, backend)
result = job.result()
#get the unitary matrix from the result object
A=result.get_unitary(qc, decimals=3);
print("Matricea unitara a circuitului 1 este \n",A)
print(qc.draw())
qreg1 = QuantumRegister(4) 
qc1 = QuantumCircuit(qreg1)
qc1.x(0)
qc1.x(2)
circuit=QuantumCircuit(4)
circuit = XGate().control(3, ctrl_state='111')
qc1.append(circuit, [0, 1, 2, 3])
circuit = XGate().control(3, ctrl_state='111')
qc1.append(circuit, [0, 1, 2, 3])
circuit = XGate().control(3, ctrl_state='111')
qc1.append(circuit, [0, 1, 2, 3])
qc1.x(0)
qc1.x(2)
job = execute(qc1, backend)
result = job.result()
#get the unitary matrix from the result object
B=result.get_unitary(qc1, decimals=3);
print("Matricea unitara a circuitului 2 este \n",B)
print(qc1.draw())
def egalitate(A,B):
    eps=sys.float_info.epsilon
    for i in range(16):
        for j in range(16):
            if((A[i][j]-B[i][j]) >= eps):
                return 0
    return 1
if(egalitate(A,B)==0):
    print("Circuitele nu sunt egale");
else:
    print("Circuitele sunt egale");
```

```python
#ex3
#a
def desenatgrafic(a,x,N):
    import matplotlib.pyplot as plt
    x1=[]
    y1=[]
    for i in range(1,x+1):
        y1.append(i)
        j=(a**i)%N;
        x1.append(j)
    plt.plot(x1, y1, label = "ex3,i")
    # naming the x axis
    plt.xlabel('x - axis')
    # naming the y axis
    plt.ylabel('y - axis')
    # function to show the plot
    plt.show()
desenatgrafic(9,100,55)
#b
def perioada(a,N):
    r=0
    j=0
    while(j!=1):
        r=r+1;
        j=(a**r)%N;
    return r;
print("perioada functiei e",perioada(9,55))
```

```python
#ex2
#versiune 1
from qiskit import QuantumCircuit,QuantumRegister
from qiskit.circuit import Parameter
from qiskit.circuit.gate import Gate
from qiskit import *
from qiskit.visualization import plot_histogram
import math
import numpy as np
pi = np.pi


circuit=QuantumCircuit(1)
teta=4*pi*0.111
circuit.ry(teta,0)
U = circuit.to_gate(label="U").control(1)

circuit=QuantumCircuit(1)
teta=4*pi*0.111
circuit.ry(teta,0)
circuit.ry(teta,0)
Ude2ori = circuit.to_gate(label="U^2").control(1)

circuit=QuantumCircuit(1)
teta=4*pi*0.111
circuit.ry(teta,0)
circuit.ry(teta,0)
circuit.ry(teta,0)
circuit.ry(teta,0)
Ude4ori = circuit.to_gate(label="U^4").control(1)

qc = QuantumCircuit(4,4)
initial_state=[1,0]
qc.initialize(initial_state,0)
qc.initialize(initial_state,1)
qc.initialize(initial_state,2)
initial_state = [0, -1j]
qc.initialize(initial_state,3)
qc.h(0)
qc.h(1)
qc.h(2)
qc.append(U, [0, 3])
qc.append(Ude2ori, [1, 3])
qc.append(Ude4ori, [2, 3])
qc.barrier()
qc.h(0)
qc.h(1)
qc.h(2)
qc.crx(-pi/4, 2, 0)
qc.crx(-pi/2, 1, 0)
qc.crx(-pi/2, 2, 1)
qc.barrier()
qc.measure([0, 1, 2], [0, 1, 2])
print(qc.draw())
simulator=Aer.get_backend('qasm_simulator')
result=execute(qc,simulator,shots=1000).result()
counts=result.get_counts()
print("Rezultatul simularii este ",counts)
plot_histogram(counts).savefig('tema2poza1.png')
```
